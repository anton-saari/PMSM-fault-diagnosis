```python
import zipfile
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, log_loss
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

zip_path = "PMSM-inverter-fault-diagnosis-3.0.zip"
extract_dir = "PMSM_inverter_fault_data"

# Extract once
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Check structure
os.listdir(extract_dir)

csv_path = "PMSM_inverter_fault_data/PMSM-inverter-fault-diagnosis-3.0/processed_data/converted_dataset.csv"
df = pd.read_csv(csv_path)
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()

y = df['FDD'].values   # fault class (F0..F8)

# Features (drop label + timestamp if not needed)
X = df.drop(columns=['FDD', 'Timestamp']).values
# Labels

FeatureCorrelation = df.drop(columns=['FDD', 'Timestamp'])
corr = FeatureCorrelation.corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", annot=False, linewidths=0.3)
plt.title("Feature Correlation Heatmap (PMSM Inverter Fault Dataset)")
plt.show()

print("Features shape:", X.shape)
print("Labels shape:", y.shape)
print("Classes:", sorted(set(y)))
print(df.head())
print("Number of features:", X.shape[1])
print("Feature names:", df.drop(columns=['FDD','Timestamp']).columns.tolist())
```

```python
le = LabelEncoder()
y_enc = le.fit_transform(y)

# 2) Train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, random_state=42, stratify=y_enc
)

# 3) Pipeline: Standardize features + Multinomial Logistic Regression (logistic loss)
#    - class_weight='balanced' helps with imbalanced classes
#    - lbfgs supports multinomial loss well; raise max_iter to ensure convergence
clf = Pipeline(steps=[
    ("scaler", StandardScaler(with_mean=True, with_std=True)),
    ("logreg", LogisticRegression(
        multi_class="multinomial",
        solver="lbfgs",
        class_weight="balanced",
        C=1.0,
        penalty="l2",
        max_iter=2000,
        n_jobs=-1,
        random_state=42
    ))
])

# 4) Fit
clf.fit(X_train, y_train)

# 5) Evaluate on test set
y_proba = clf.predict_proba(X_test)
y_pred  = clf.predict(X_test)

acc = accuracy_score(y_test, y_pred)
f1m = f1_score(y_test, y_pred, average="macro")
ll  = log_loss(y_test, y_proba, labels=list(range(len(le.classes_))))

print(f"Test accuracy: {acc:.4f}")
print(f"Macro F1     : {f1m:.4f}")
print(f"Log loss     : {ll:.4f}")
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# 6) Confusion matrix
fig, ax = plt.subplots(figsize=(6,5))
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    labels=np.arange(len(le.classes_)),   # your y_* are encoded ints 0..K-1
    display_labels=le.classes_,           # show "F0".."F8" on axes
    normalize='true',                     # per-row normalization
    ax=ax
)
ax.set_title("Normalized confusion matrix")
ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
plt.tight_layout()
plt.show()
# 7) Optional: 5-fold CV with multiple metrics (on the whole dataset)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_validate(
    clf, X, y_enc, cv=cv,
    scoring={"acc": "accuracy", "f1_macro": "f1_macro", "log_loss": "neg_log_loss"},
    n_jobs=-1, return_train_score=False
)
print("\n5-fold CV (mean  std):")
print(f"Accuracy : {cv_scores['test_acc'].mean():.4f}  {cv_scores['test_acc'].std():.4f}")
print(f"F1_macro : {cv_scores['test_f1_macro'].mean():.4f}  {cv_scores['test_f1_macro'].std():.4f}")
print(f"Log loss : {-cv_scores['test_log_loss'].mean():.4f}  {cv_scores['test_log_loss'].std():.4f}")
```
